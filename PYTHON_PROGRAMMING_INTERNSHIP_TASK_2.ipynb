## **2.3. Getting User Input for Portfolio**
 2.3.1 Helper: get positive integer

 2.3.2 Helper: display numbered stock list and get selection

 2.3.3 Main function to collect portfolio (using numbered menu)

 2.3.4 test block – not executed automatically

````python
# ==================================================
# 2.3. Getting User Input for Portfolio
# ==================================================

# ----------------------------
# 2.3.1 Helper: get positive integer
# ----------------------------
def get_positive_integer(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value > 0:
                return value
            else:
                print_colored("❌ Please enter a positive number.", 'red')
        except ValueError:
            print_colored("❌ Invalid input. Enter a whole number.", 'red')

# ----------------------------
# 2.3.2 Helper: display numbered stock list and get selection
# ----------------------------
def get_stock_choice():
    """
    Display available stocks with numbers, let user choose by number.
    Returns the stock symbol (e.g., 'AAPL').
    """
    # Create a numbered list from STOCK_PRICES items
    stock_items = list(STOCK_PRICES.items())  # list of (symbol, price)
    print_colored("\n📋 Available stocks:", 'cyan')
    for idx, (symbol, price) in enumerate(stock_items, start=1):
        print(f"   {idx}. {symbol} - {format_currency(price)}")

    while True:
        try:
            choice = int(input("\nSelect stock by number: "))
            if 1 <= choice <= len(stock_items):
                symbol = stock_items[choice-1][0]
                return symbol
            else:
                print_colored(f"❌ Please choose a number between 1 and {len(stock_items)}.", 'red')
        except ValueError:
            print_colored("❌ Invalid input. Enter a number.", 'red')

# ----------------------------
# 2.3.3 Main function to collect portfolio (using numbered menu)
# ----------------------------
def get_user_portfolio():
    portfolio = []
    print_colored("\n📝 Let's build your portfolio.", 'cyan', bold=True)

    num_stocks = get_positive_integer("How many different stocks do you own? ")

    for i in range(num_stocks):
        print_colored(f"\n--- Stock #{i+1} ---", 'yellow')
        symbol = get_stock_choice()                     # choose by number
        shares = get_positive_integer(f"Enter number of shares for {symbol}: ")
        price = get_stock_price(symbol)
        value = price * shares
        portfolio.append({
            'symbol': symbol,
            'shares': shares,
            'price': price,
            'value': value
        })
        print_colored(f"✅ Added {shares} shares of {symbol} @ {format_currency(price)} = {format_currency(value)}", 'green')

    return portfolio

# ----------------------------
# 2.3.4  test block – not executed automatically
# ----------------------------
if __name__ == "__main__":
    test_portfolio = get_user_portfolio()
    print("\n📦 Collected portfolio:")
    for item in test_portfolio:
        print(item)
````

## **2.5. Saving Results to a File**
 2.5.1 Function to save portfolio as CSV

 2.5.2 Function to save portfolio as TXT (human‑readable)

 2.5.3 Function to ask user about saving and handle download

 2.5.4 Integration with existing workflow (call after displaying portfolio)


````python
# ==================================================
# 2.5. Saving Results to a File
# ==================================================

import csv
from datetime import datetime
from google.colab import files

# ----------------------------
# 2.5.1 Function to save portfolio as CSV
# ----------------------------
def save_portfolio_csv(portfolio, total, filename=None):
    """
    Save portfolio summary to a CSV file.

    Parameters:
    portfolio (list): List of holdings (dict with keys 'symbol', 'shares', 'price', 'value')
    total (float): Total portfolio value
    filename (str): Optional custom filename. If None, auto-generate with timestamp.

    Returns:
    str: The filename used.
    """
    if not portfolio:
        print_colored("⚠️ Nothing to save – portfolio is empty.", 'yellow')
        return None

    if filename is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"portfolio_{timestamp}.csv"

    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['Symbol', 'Shares', 'Price', 'Value']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for item in portfolio:
            writer.writerow({
                'Symbol': item['symbol'],
                'Shares': item['shares'],
                'Price': f"{item['price']:.2f}",
                'Value': f"{item['value']:.2f}"
            })
        # Write total row
        writer.writerow({})
        writer.writerow({'Symbol': 'TOTAL', 'Shares': '', 'Price': '', 'Value': f"{total:.2f}"})

    print_colored(f"✅ Portfolio saved to {filename}", 'green')
    return filename

# ----------------------------
# 2.5.2 Function to save portfolio as TXT (human‑readable)
# ----------------------------
def save_portfolio_txt(portfolio, total, filename=None):
    """Save portfolio as a formatted text file."""
    if not portfolio:
        print_colored("⚠️ Nothing to save – portfolio is empty.", 'yellow')
        return None

    if filename is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"portfolio_{timestamp}.txt"

    with open(filename, 'w') as f:
        f.write("STOCK PORTFOLIO SUMMARY\n")
        f.write(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("=" * 50 + "\n")
        f.write(f"{'Symbol':<8} {'Shares':>10} {'Price':>12} {'Value':>14}\n")
        f.write("-" * 50 + "\n")
        for item in portfolio:
            f.write(f"{item['symbol']:<8} {item['shares']:>10} ${item['price']:>10.2f} ${item['value']:>13.2f}\n")
        f.write("-" * 50 + "\n")
        f.write(f"{'TOTAL':<8} {'':>10} {'':>12} ${total:>13.2f}\n")

    print_colored(f"✅ Portfolio saved to {filename}", 'green')
    return filename

# ----------------------------
# 2.5.3 Function to ask user about saving and handle download
# ----------------------------
def offer_save_portfolio(portfolio, total):
    """Ask user if they want to save, offer format, and download."""
    if not portfolio:
        return

    print_colored("\n💾 Do you want to save your portfolio?", 'cyan', bold=True)
    choice = input("Save to file? (y/n): ").strip().lower()
    if choice not in ['y', 'yes']:
        print_colored("Skipped saving.", 'yellow')
        return

    print("Choose format:")
    print("   1. CSV (spreadsheet)")
    print("   2. TXT (readable text)")
    format_choice = input("Enter 1 or 2: ").strip()

    if format_choice == '1':
        filename = save_portfolio_csv(portfolio, total)
    elif format_choice == '2':
        filename = save_portfolio_txt(portfolio, total)
    else:
        print_colored("Invalid choice. Saving as CSV by default.", 'yellow')
        filename = save_portfolio_csv(portfolio, total)

    if filename:
        # Download file in Colab
        try:
            files.download(filename)
            print_colored(f"📥 Download started: {filename}", 'green')
        except:
            print_colored("⚠️ Could not trigger download automatically. File is saved in the Colab environment.", 'yellow')

# ----------------------------
# 2.5.4 Integration with existing workflow (call after displaying portfolio)
# ----------------------------
# Assuming `portfolio` and `total` are already defined from Subtask 4
# You can call this function at the end of your main program:
# offer_save_portfolio(portfolio, total)
````

## **2.6. Adding, Polishing & User Experience**

 2.6.4 Improved portfolio display with better formatting

 2.6.6 Main program with menu loop

 2.6.7 Run the polished tracker


````python
# ==================================================
# 2.6. Adding Polishing & User Experience
# ==================================================

import sys
import time

# ----------------------------
# 2.6.1 Clear screen simulation (works in Colab)
# ----------------------------
def clear_screen():
    """Print newlines to simulate screen clearing."""
    print("\n" * 2)

# ----------------------------
# 2.6.2 Pause and wait for user to press Enter
# ----------------------------
def wait_for_enter():
    input("\n⏎ Press Enter to continue...")

# ----------------------------
# 2.6.3 Main menu display
# ----------------------------
def show_main_menu():
    """Display the main menu options."""
    print_colored("\n" + "=" * 50, 'cyan', bold=True)
    print_colored("💰 STOCK PORTFOLIO TRACKER", 'green', bold=True)
    print_colored("=" * 50, 'cyan', bold=True)
    print("1. Enter / Update Portfolio")
    print("2. View Current Portfolio")
    print("3. Save Portfolio to File")
    print("4. Exit")
    print("-" * 50)

# ----------------------------
# 2.6.4 Improved portfolio display with better formatting
# ----------------------------
def display_portfolio_polished(portfolio, total):
    """Display portfolio with better visual separation."""
    if not portfolio:
        print_colored("\n⚠️ Portfolio is empty. Please add holdings first.", 'yellow')
        return False

    print_colored("\n📊 YOUR PORTFOLIO SUMMARY", 'cyan', bold=True)
    print(f"{'Symbol':<8} {'Shares':>10} {'Price':>12} {'Value':>14}")
    print("-" * 50)
    for item in portfolio:
        print(f"{item['symbol']:<8} {item['shares']:>10} {format_currency(item['price']):>12} {format_currency(item['value']):>14}")
    print("-" * 50)
    print_colored(f"{'TOTAL':<8} {'':>10} {'':>12} {format_currency(total):>14}", 'green', bold=True)

    # Show summary insights
    print_colored("\n📈 INVESTMENT SUMMARY", 'purple', bold=True)
    print(f"   Number of holdings: {len(portfolio)}")
    print(f"   Total invested value: {format_currency(total)}")
    largest = max(portfolio, key=lambda x: x['value'])
    smallest = min(portfolio, key=lambda x: x['value'])
    print(f"   Largest holding: {largest['symbol']} ({format_currency(largest['value'])})")
    print(f"   Smallest holding: {smallest['symbol']} ({format_currency(smallest['value'])})")
    return True

# ----------------------------
# 2.6.5 Validate portfolio before saving
# ----------------------------
def validate_portfolio_for_saving(portfolio):
    """Check if portfolio exists and is non-empty."""
    if not portfolio:
        print_colored("❌ Cannot save – portfolio is empty. Please add stocks first.", 'red')
        return False
    return True

# ----------------------------
# 2.6.6 Main program with menu loop
# ----------------------------
def run_stock_tracker():
    """Main interactive loop for the portfolio tracker."""
    portfolio = []      # will be filled by user
    total = 0.0

    while True:
        clear_screen()
        show_main_menu()
        choice = input("👉 Enter your choice (1-4): ").strip()

        if choice == '1':
            # Enter / Update Portfolio
            portfolio = get_user_portfolio()   # from Subtask 3
            total = sum(item['value'] for item in portfolio)
            print_colored("\n✅ Portfolio updated successfully!", 'green')
            wait_for_enter()

        elif choice == '2':
            # View Current Portfolio
            if portfolio:
                clear_screen()
                display_portfolio_polished(portfolio, total)
            else:
                print_colored("\n📭 No portfolio data yet. Please add stocks first (option 1).", 'yellow')
            wait_for_enter()

        elif choice == '3':
            # Save Portfolio to File
            if validate_portfolio_for_saving(portfolio):
                offer_save_portfolio(portfolio, total)   # from Subtask 5
            wait_for_enter()

        elif choice == '4':
            print_colored("\n👋 Thank you for using Stock Portfolio Tracker!", 'cyan', bold=True)
            print_colored("   Have a great day!\n", 'green')
            break

        else:
            print_colored("❌ Invalid choice. Please enter 1, 2, 3, or 4.", 'red')
            wait_for_enter()

# ----------------------------
# 2.6.7 Run the polished tracker
# ----------------------------
if __name__ == "__main__":
    # Ensure all required functions are defined before running
    try:
        run_stock_tracker()
    except KeyboardInterrupt:
        print_colored("\n\n👋 Program interrupted. Goodbye!", 'yellow')
````

````python
# ==================================================
# 2.7. Saving & Documenting the Project
# ==================================================

import os
import zipfile
from datetime import datetime
from google.colab import files

# Create results directory
if not os.path.exists('stock_tracker_results'):
    os.makedirs('stock_tracker_results')
print("📁 Created directory: stock_tracker_results/")

# ----------------------------
# 2.7.1 Save the complete script as a standalone Python file
# ----------------------------
script_content = '''# Stock Portfolio Tracker – CodeAlpha Task 2
# A console-based tool to calculate and manage stock portfolio values.

import csv
from datetime import datetime

# ---------- Coloured output ----------
def print_colored(text, color='white', bold=False):
    colors = {
        'red': '\033[91m', 'green': '\033[92m', 'yellow': '\033[93m',
        'blue': '\033[94m', 'purple': '\033[95m', 'cyan': '\033[96m',
        'white': '\033[97m', 'reset': '\033[0m'
    }
    bold_code = '\033[1m' if bold else ''
    reset_code = colors['reset']
    color_code = colors.get(color, colors['white'])
    print(f"{bold_code}{color_code}{text}{reset_code}")

# ---------- Format currency ----------
def format_currency(value):
    return f"${value:,.2f}"

# ---------- Stock data ----------
STOCK_PRICES = {
    "AAPL": 180.25, "TSLA": 250.50, "GOOGL": 140.30,
    "MSFT": 330.00, "AMZN": 128.50, "META": 310.75
}

def is_valid_symbol(symbol):
    return symbol.upper() in STOCK_PRICES

def get_stock_price(symbol):
    return STOCK_PRICES.get(symbol.upper(), None)

def display_available_stocks():
    print_colored("\n📋 AVAILABLE STOCKS", 'cyan', bold=True)
    print(f"{'Symbol':<8} {'Price':>12}")
    print("-" * 22)
    for sym, price in STOCK_PRICES.items():
        print(f"{sym:<8} {format_currency(price):>12}")
    print()

# ---------- Helper input functions ----------
def get_positive_integer(prompt):
    while True:
        try:
            val = int(input(prompt))
            if val > 0:
                return val
            print_colored("❌ Please enter a positive number.", 'red')
        except ValueError:
            print_colored("❌ Invalid input. Enter a whole number.", 'red')

def display_stock_menu():
    items = list(STOCK_PRICES.items())
    for idx, (sym, price) in enumerate(items, 1):
        print(f"   {idx}. {sym} – {format_currency(price)}")
    return items

def get_symbol_by_choice():
    items = display_stock_menu()
    while True:
        try:
            choice = int(input("\n👉 Select stock by number: "))
            if 1 <= choice <= len(items):
                return items[choice-1][0]
            print_colored(f"❌ Enter a number between 1 and {len(items)}.", 'red')
        except ValueError:
            print_colored("❌ Invalid input. Enter a number.", 'red')

def get_user_portfolio():
    portfolio = []
    print_colored("\n📝 Let's build your portfolio.", 'cyan', bold=True)
    num = get_positive_integer("How many different stocks do you own? ")
    for i in range(num):
        print_colored(f"\n--- Stock #{i+1} ---", 'yellow')
        symbol = get_symbol_by_choice()
        shares = get_positive_integer(f"Enter number of shares for {symbol}: ")
        price = get_stock_price(symbol)
        value = price * shares
        portfolio.append({'symbol': symbol, 'shares': shares, 'price': price, 'value': value})
        print_colored(f"✅ Added {shares} shares of {symbol} @ {format_currency(price)} = {format_currency(value)}", 'green')
    return portfolio

# ---------- Portfolio display ----------
def display_portfolio(portfolio, total):
    if not portfolio:
        print_colored("\n⚠️ Portfolio is empty.", 'yellow')
        return
    print_colored("\n📊 YOUR PORTFOLIO SUMMARY", 'cyan', bold=True)
    print(f"{'Symbol':<8} {'Shares':>10} {'Price':>12} {'Value':>14}")
    print("-" * 50)
    for item in portfolio:
        print(f"{item['symbol']:<8} {item['shares']:>10} {format_currency(item['price']):>12} {format_currency(item['value']):>14}")
    print("-" * 50)
    print_colored(f"{'TOTAL':<8} {'':>10} {'':>12} {format_currency(total):>14}", 'green', bold=True)

def portfolio_summary(portfolio, total):
    if not portfolio:
        return
    print_colored("\n📈 INVESTMENT SUMMARY", 'purple', bold=True)
    print(f"   Number of holdings: {len(portfolio)}")
    print(f"   Total invested value: {format_currency(total)}")
    largest = max(portfolio, key=lambda x: x['value'])
    smallest = min(portfolio, key=lambda x: x['value'])
    print(f"   Largest holding: {largest['symbol']} ({format_currency(largest['value'])})")
    print(f"   Smallest holding: {smallest['symbol']} ({format_currency(smallest['value'])})")

# ---------- File saving ----------
def save_portfolio_csv(portfolio, total, filename=None):
    if not portfolio:
        return None
    if filename is None:
        filename = f"portfolio_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
    with open(filename, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['Symbol', 'Shares', 'Price', 'Value'])
        writer.writeheader()
        for item in portfolio:
            writer.writerow({'Symbol': item['symbol'], 'Shares': item['shares'],
                             'Price': f"{item['price']:.2f}", 'Value': f"{item['value']:.2f}"})
        writer.writerow({})
        writer.writerow({'Symbol': 'TOTAL', 'Shares': '', 'Price': '', 'Value': f"{total:.2f}"})
    print_colored(f"✅ Portfolio saved to {filename}", 'green')
    return filename

def offer_save_portfolio(portfolio, total):
    if not portfolio:
        print_colored("❌ Nothing to save.", 'red')
        return
    choice = input("\nSave portfolio? (y/n): ").strip().lower()
    if choice not in ['y', 'yes']:
        return
    filename = save_portfolio_csv(portfolio, total)
    if filename:
        try:
            from google.colab import files
            files.download(filename)
            print_colored("📥 Download started.", 'green')
        except:
            print_colored("File saved in environment.", 'yellow')

# ---------- Main interactive loop ----------
def main():
    print_colored("\n💰 STOCK PORTFOLIO TRACKER", 'green', bold=True)
    portfolio = []
    total = 0
    while True:
        print_colored("\n" + "="*40, 'cyan')
        print("1. Enter/Update Portfolio")
        print("2. View Portfolio")
        print("3. Save Portfolio")
        print("4. Exit")
        choice = input("👉 Choose (1-4): ").strip()
        if choice == '1':
            portfolio = get_user_portfolio()
            total = sum(item['value'] for item in portfolio)
            print_colored("✅ Portfolio updated.", 'green')
        elif choice == '2':
            display_portfolio(portfolio, total)
            portfolio_summary(portfolio, total)
        elif choice == '3':
            offer_save_portfolio(portfolio, total)
        elif choice == '4':
            print_colored("👋 Goodbye!", 'cyan')
            break
        else:
            print_colored("❌ Invalid choice.", 'red')
        input("\n⏎ Press Enter to continue...")

if __name__ == "__main__":
    main()
'''

with open('stock_tracker_results/stock_portfolio_tracker.py', 'w') as f:
    f.write(script_content)
print("💾 Saved standalone Python script: stock_portfolio_tracker.py")

# ----------------------------
# 2.7.2 Generate README.md for GitHub (fixed string)
# ----------------------------
readme_content = (
    "# Stock Portfolio Tracker – CodeAlpha Task 2\n\n"
    "## 📌 Project Overview\n"
    "A console-based Python application that helps users calculate the total value of their stock portfolio using hardcoded stock prices. "
    "Users select stocks from a numbered menu, enter quantities, and the program displays a formatted portfolio summary with total value. "
    "Optionally, the portfolio can be saved as a CSV file.\n\n"
    "## 🎮 Features\n"
    "- Predefined stock prices (AAPL, TSLA, GOOGL, MSFT, AMZN, META)\n"
    "- Numbered menu for stock selection (no typing errors)\n"
    "- Input validation (positive shares, valid selections)\n"
    "- Coloured console output (green for success, red for errors)\n"
    "- Formatted table display with currency formatting\n"
    "- Investment summary (largest/smallest holding)\n"
    "- Save portfolio to CSV and download in Colab\n"
    "- Interactive menu loop (enter, view, save, exit)\n\n"
    "## 🛠️ Tech Stack\n"
    "- Python 3.x (standard library only – no external dependencies)\n"
    "- Google Colab / any terminal with Python\n\n"
    "## 🚀 How to Use\n"
    "1. Run the script: `python stock_portfolio_tracker.py`\n"
    "2. Choose option 1 to enter your portfolio.\n"
    "3. Select stocks by number, enter shares.\n"
    "4. View portfolio (option 2) to see table and total.\n"
    "5. Save to CSV (option 3) and download.\n\n"
    "## 📁 Repository Structure\n"
    "```\n"
    "CodeAlpha_Stock_Portfolio_Tracker/\n"
    "├── stock_portfolio_tracker.py   # Main script\n"
    "├── Stock_Portfolio_Tracker.ipynb # Colab notebook (if used)\n"
    "├── README.md                    # This file\n"
    "├── requirements.txt             # (empty, no dependencies)\n"
    "└── stock_tracker_results/       # Output folder (optional)\n"
    "```\n\n"
    "## 🖼️ Sample Output\n"
    "```\n"
    "📊 YOUR PORTFOLIO SUMMARY\n"
    "Symbol    Shares       Price         Value\n"
    "--------------------------------------------------\n"
    "AAPL          10     $180.25     $1,802.50\n"
    "TSLA           5     $250.50     $1,252.50\n"
    "MSFT           2     $330.00       $660.00\n"
    "TOTAL                         $3,715.00\n"
    "```\n\n"
    "## 📽️ Video Explanation\n"
    "[Link to LinkedIn video] – Replace with your actual URL.\n\n"
    "## 👩‍💻 Author\n"
    "Your Name – [LinkedIn Profile](https://linkedin.com/in/yourprofile)\n\n"
    "## 🙏 Acknowledgements\n"
    "CodeAlpha for the internship opportunity.\n\n"
    "---\n"
    "⭐ Star this repo if you found it useful!\n"
)

with open('README.md', 'w') as f:
    f.write(readme_content)
print("📝 README.md created for GitHub")

# ----------------------------
# 2.7.3 Create requirements.txt
# ----------------------------
with open('requirements.txt', 'w') as f:
    f.write("# No external dependencies required for Stock Portfolio Tracker\n# Python standard library only\n")
print("📦 requirements.txt created")

# ----------------------------
# 2.7.4 Create a sample output file (optional)
# ----------------------------
sample_output = """Sample portfolio output (example):
Symbol    Shares       Price         Value
AAPL          10     $180.25     $1,802.50
TSLA           5     $250.50     $1,252.50
MSFT           2     $330.00       $660.00
TOTAL                         $3,715.00
"""
with open('stock_tracker_results/sample_output.txt', 'w') as f:
    f.write(sample_output)
print("📄 Sample output saved for reference")

# ----------------------------
# 2.7.5 Zip all results for download
# ----------------------------
zip_path = 'stock_portfolio_tracker_results.zip'
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, dirs, files_in_dir in os.walk('stock_tracker_results'): # Renamed 'files' to 'files_in_dir'
        for file in files_in_dir:
            zipf.write(os.path.join(root, file), arcname=os.path.join('stock_tracker_results', file))
    zipf.write('README.md')
    zipf.write('requirements.txt')
    if os.path.exists('Stock_Portfolio_Tracker.ipynb'):
        zipf.write('Stock_Portfolio_Tracker.ipynb')

print(f"🗜️ Zipped all results to {zip_path}")
files.download(zip_path)
print("✅ Download started. Your results zip file is being saved locally.")

# ----------------------------
# 2.7.6 Final Completion Message
# ----------------------------
print("\n" + "="*60)
print("🎉 TASK 2 (STOCK PORTFOLIO TRACKER) COMPLETED SUCCESSFULLY! 🎉")
print("="*60)
print("All results saved in 'stock_tracker_results/' folder and zipped for download.")
print("\n📌 Next Steps for Submission:")
print("1. Upload the notebook/script to GitHub repository: CodeAlpha_Stock_Portfolio_Tracker")
print("2. Add README.md and requirements.txt to the repo")
print("3. Add the 'stock_tracker_results' folder contents (or at least the sample output)")
print("4. Record a 1-2 minute video demonstrating the tool (menu, input, table, save)")
print("5. Post video on LinkedIn with @CodeAlpha tag and GitHub link")
print("6. Submit the form shared in the WhatsApp group")
print("\n🌟 Great work on completing Task 2!")
````
